```python
def feat_plot(attribute):
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    attribute.plot(kind = 'hist')
    plt.title(f'{attribute.name} histogram plot')

    #mean = feature.describe().mean()
    plt.subplot(1, 3, 2)
    mu, sigma = scipy.stats.norm.fit(attribute)
    sns.distplot(attribute)
    #plt.legend({'--': mu, 'sigma': sigma})
    plt.axvline(mu, linestyle = '--', color = 'green', )
    plt.axvline(sigma, linestyle = '--', color = 'red')
    plt.title(f'{attribute.name} distribution plot')

    plt.subplot(1, 3, 3)
    sns.boxplot(attribute)
    plt.title(f'{attribute.name} box plot')
    plt.show()
```

```python
num_feat = df1.select_dtypes(exclude = 'object')
for i in num_feat.columns:
    feat_plot(num_feat[i])
```
